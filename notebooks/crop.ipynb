```python
def crop(Bbox, image):
    im=plt.imread(image)
    y= im.shape[0]
    x= im.shape[1]
    x1=round(Bbox[0]*x)
    y1=round(Bbox[1]*y)
    width=round(Bbox[2]*x)
    height=round(Bbox[3]*y)
    #print(x1, y1, width, height)
    x2, y2 = x1 + width, y1 + height
    new_image = im[y1:y2, x1:x2]
    plt.imshow(new_image)
    plt.savefig('./test_crop/'+image)
```
